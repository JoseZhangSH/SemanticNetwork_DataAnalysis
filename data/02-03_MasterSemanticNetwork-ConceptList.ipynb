```python
# 数据处理
import pandas as pd
import numpy as np
import json
import csv
import random
import math
from itertools import product


# 网络分析与可视化
import networkx as nx
from networkx.readwrite import json_graph
from pyvis import network as net
import matplotlib.pyplot as plt
import powerlaw # Power laws are probability distributions with the form:p(x)∝x−α
```

```python
with open('02_Graph/MasterConceptNetwork_Word2Vec-0.62_Baseline-Test_0331.json') as f:
    js_graph = json.load(f)
    G_MasterConceptNetwork = json_graph.node_link_graph(js_graph)

df_concept_info = pd.read_csv('00_Raw Data/CCFD_concept_info.csv')
```

```python
# 查看全部节点在某个属性的所有选项
def NodeAttributeValueList(G,attribute):
    return set(np.array([G.nodes[n][attribute] for n in G.nodes]).flatten())
```

```python
maincategorys = NodeAttributeValueList(G_MasterConceptNetwork,'maincategory')
```

```python
ConceptList = []

for i in maincategorys:
    ConceptList.append({
        "title":i,
        "key":i,
        "children":[]
    })

for node in G_MasterConceptNetwork.nodes():
    if('picture_naming_status' in G_MasterConceptNetwork.nodes[node].keys()):
        picture_naming_status = "true"
        if(G_MasterConceptNetwork.nodes[node]['picture_naming_result']):
            picture_naming_result = "true"
        else:
            picture_naming_result = "false"
    else:
        picture_naming_status = "false"
        picture_naming_result = "false"
    
    iconKey = G_MasterConceptNetwork.nodes[node]['maincategory']+"-"+picture_naming_status+"-"+picture_naming_result


    [item for item in ConceptList if item.get('title')==G_MasterConceptNetwork.nodes[node]['maincategory']][0]['children'].append({
        "title":node,
        "key":node,
        "slots":{
            "icon": iconKey
        }
    })
```

```python
for i in ConceptList:
    i['children'].sort(key=lambda x:G_MasterConceptNetwork.nodes[x['title']]['degree_centrality'],reverse=True)
```
